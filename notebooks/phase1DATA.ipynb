```python
danger_zones['Wilaya'] = algerian_communes['Wilaya']
danger_zones['Level'] = algerian_communes['Danger_level']
```

```python
color_map = {
    'Zone 0': 'green',
    'Zone I': 'yellow',
    'Zone II': 'orange',
    'Zone IIa': 'orange',
    'Zone IIb': 'orange',
    'Zone III': 'red'
}
```

```python
m_algeria = folium.Map(location=[28.0339, 1.6596], tiles='cartodbpositron', zoom_start=6)
```

```python
for idx, row in danger_zones.iterrows():
    # Extract latitude and longitude from the POINT geometry
    lat = row.geometry.y
    lon = row.geometry.x
    
    folium.CircleMarker(
        location=[lat, lon],
        radius=10, # Size of the bubble
        color=color_map.get(row['Level'], 'gray'),
        fill=True,
        fill_opacity=0.7,
        popup=f"Wilaya: {row['Wilaya']}<br>Danger: {row['Level']}"
    ).add_to(m_6)
```

```python
sismic_zones = danger_zones
```

```python
sismic_zones[['Lon', 'Lat']] = sismic_zones['geometry'].astype(str).str.extract(r'\(([-?0-9.]+) ([-?0-9.]+)\)')
sismic_zones[['Lon', 'Lat']] = sismic_zones[['Lon', 'Lat']].astype(float)
```

```python
sismic_zones.head()
```

```python
m_2 = folium.Map(location=[32.0339, 1.6596], tiles='cartodbpositron', zoom_start=6)
```

```python

weights = {'Zone 0': 0.1, 'Zone I': 0.2, 'Zone IIa': 0.4, 'Zone IIb': 0.7, 'Zone III': 0.9}
sismic_zones['Weight'] = sismic_zones['Level'].map(weights)

from folium.plugins import HeatMap

data = sismic_zones[['Lat', 'Lon', 'Weight']].values.tolist()
HeatMap(data, radius=15, blur=10).add_to(m_2)
```
